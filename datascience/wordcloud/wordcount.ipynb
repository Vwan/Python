```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

filename = "words.txt"
with open(filename) as f:
    mytext = f.read()
mytext

wordcloud = WordCloud().generate(mytext)

%pylab inline

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
